## dsub test

### Job script

```python
%%writefile kir_dsub_test.sh

#!/bin/bash

# select a region and convert cram to bam
# this is without looping, using 3 participants to test scaling up to 10/VM; try looping for large scale
# 01
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_1 \
--read-index $INPUT_CRAI_1 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample01_chr19_region.bam
        
# 02
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_2 \
--read-index $INPUT_CRAI_2 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample02_chr19_region.bam
        
# 03
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_3 \
--read-index $INPUT_CRAI_3 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample03_chr19_region.bam
        
# 04
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_1 \
--read-index $INPUT_CRAI_1 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample04_chr19_region.bam
        
# 05
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_2 \
--read-index $INPUT_CRAI_2 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample05_chr19_region.bam
        
# 06
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_3 \
--read-index $INPUT_CRAI_3 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample06_chr19_region.bam
        
# 07
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_1 \
--read-index $INPUT_CRAI_1 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample07_chr19_region.bam
        
# 08
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_2 \
--read-index $INPUT_CRAI_2 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample08_chr19_region.bam
        
# 09
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_3 \
--read-index $INPUT_CRAI_3 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample09_chr19_region.bam
        
# 10
gatk PrintReads \
-R gs://gcp-public-data--broad-references/hg38/v0/Homo_sapiens_assembly38.fasta \
-I $INPUT_CRAM_1 \
--read-index $INPUT_CRAI_1 \
-L chr19:54000000-55100000 \
--gcs-project-for-requester-pays $GOOGLE_PROJECT \
-O sample10_chr19_region.bam
        
            
# run kir-mapper
# Step 1: Map (using your BAM file)
kir-mapper map -bam sample01_chr19_region.bam -sample sample01 -output ./kir_output -threads 3
kir-mapper map -bam sample02_chr19_region.bam -sample sample02 -output ./kir_output -threads 3
kir-mapper map -bam sample03_chr19_region.bam -sample sample03 -output ./kir_output -threads 3
kir-mapper map -bam sample04_chr19_region.bam -sample sample04 -output ./kir_output -threads 3
kir-mapper map -bam sample05_chr19_region.bam -sample sample05 -output ./kir_output -threads 3
kir-mapper map -bam sample06_chr19_region.bam -sample sample06 -output ./kir_output -threads 3
kir-mapper map -bam sample07_chr19_region.bam -sample sample07 -output ./kir_output -threads 3
kir-mapper map -bam sample08_chr19_region.bam -sample sample08 -output ./kir_output -threads 3
kir-mapper map -bam sample09_chr19_region.bam -sample sample09 -output ./kir_output -threads 3
kir-mapper map -bam sample10_chr19_region.bam -sample sample10 -output ./kir_output -threads 3

# subsequent steps can be done with 1 single command for all samples
# Step 2: Copy number estimation
kir-mapper ncopy -output ./kir_output -threads 3

# Step 3: Genotype (SNP and allele calling)
kir-mapper genotype -output ./kir_output -threads 3

# Step 4: Haplotype (requires â‰¥50 samples for accuracy)
kir-mapper haplotype -output ./kir_output -threads 3

# Copy kir output folder to dsub output folder
cp -r kir_output/* $OUTPUT_DIR
```

### Run dsub

```python
kir_dsub = at.Dsub(
    provider="google-batch",
    machine_type="c2d-highcpu-4",
    docker_image="phetk/gatk-kirmapper:0.1",
    job_script_name="kir_dsub_test.sh",
    job_name="kir_dsub",
    env_dict={
        "GOOGLE_PROJECT": os.getenv("GOOGLE_PROJECT"),
        "INPUT_CRAM_1": "gs://fc-aou-datasets-controlled/pooled/wgs/cram/v8_delta/wgs_1000000.cram",
        "INPUT_CRAM_2": "gs://fc-aou-datasets-controlled/pooled/wgs/cram/v6_base/wgs_1000004.cram",
        "INPUT_CRAM_3": "gs://fc-aou-datasets-controlled/pooled/wgs/cram/v6_base/wgs_1000033.cram"
    },
    input_dict={
        "INPUT_CRAI_1": "gs://fc-aou-datasets-controlled/pooled/wgs/cram/v8_delta/wgs_1000000.cram.crai",
        "INPUT_CRAI_2": "gs://fc-aou-datasets-controlled/pooled/wgs/cram/v6_base/wgs_1000004.cram.crai",
        "INPUT_CRAI_3": "gs://fc-aou-datasets-controlled/pooled/wgs/cram/v6_base/wgs_1000033.cram.crai"
    },
    output_dict={},
    custom_args=f"--output-recursive OUTPUT_DIR={bucket}/dsub/results/kir/"
)
kir_dsub.run(show_command=True)
```
